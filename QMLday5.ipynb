```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] =(20.0,10.0)
#rcparams refer to the runtime configuration parameters in matplotlib. These parametrs allows you to customize the default settings for various properties of matplotlib plots such as figure size,line styles,colours and more.
#figure.figsize controls the default size of the plots. 20 is the width and 10 is the height.
```

```python
# to read the data

data= pd.read_csv('/content/headbrain.csv')
print(data.shape)  #prints the numbers of rows and columns
```

```python
data.head()  #displays a sample table
```

```python
# to find the best fit line, we are to collect the values of X and Y as follows:

X=data['Head Size(cm^3)'].values
print(X)
print()
Y=data['Brain Weight(grams)'].values
print(Y)
```

```python
# in order to perform the mathematics of linear regression, we got to know the mean value of X and Y:
mean_x=np.mean(X)
print(mean_x)
mean_y=np.mean(Y)
print(mean_y)
```

```python
# In the equation of the regression line (y=mx+c), m has the following formula:  m=Σ(x-x_mean)(y-y_mean)/(x-x_mean)**2
num=0
den=0
h=len(X)
for i in range(h):
  num += (X[i] - mean_x)* (Y[i] - mean_y)
  den += (X[i] - mean_x) ** 2
m=num/den
print(m)
```

```python
# to find the c intercept now: c = y_mean-m(x_mean):
c=mean_y-(m*mean_x)
print(c)
```

```python
#Now for the given values of m and c we shall find the values of y (predicted values) and obtain the new:  Y_predicted.
#when multiplying and adding some contant to a list, the output is again a list.

Y_predicted= m*X+c
print(Y_predicted)
```

```python
#plotting the graph:

plt.scatter(X,Y,label='scatter plot') #Only when labels are specificed we get the function plt.legend() executed.
plt.plot(X,Y_predicted,color='red',label='regression line')
plt.legend()
plt.xlabel('Head size in cm^3')
plt.ylabel('Brain Weight in grams')
plt.title('Head Size vs Brain Weight')
plt.show()
```

```python
#calculation of R**2 value: R-squared value is a statistical measure of how close the data are to the fitted regression line. it is also called as the coefficient ot determination.
#R**2=R= Σ(Y_predicted-mean_y)**2/ Σ(y-mean_y)**2

Num=0
Den=0
for i in range(h):
  Num+= (Y_predicted[i]-mean_y)**2
  Den+= (Y[i]-mean_y)**2
R=Num/Den
print(R)
```

```python
# Regression analysis: is a form of predictive modeling technique which investigates the relationship betweeen the dependent and independent variables.
# In linear regression the eqn. of line used is y=mx+c where y is the dependent variable and x is the independent variable.
# 3 major uses of regression analysis : Determining the strength of the predictors, Forecasting an effort and trend forecasting
#Goal is to reduce the error and the error is stated as the difference between the estimated value and the actual value.
#The best fit line has the least distance betweeen the estimated and the actual value.
#The mathematical model is based on the principal of least square method.
```
